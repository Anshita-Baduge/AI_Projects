```python
# SMART STUDY PLANNER â€“ AI PROJECT

# UNIT-I: Agent, Environment, State Space

subjects = {
    "Deep Learning": 3,
    "Artificial Intelligence": 4,
    "Compiler Design": 2

}

time_slots = ["Morning", "Afternoon", "Evening"]

# UNIT-II: Uninformed Search
def generate_plans(subjects, slots):
    plans = []
    s = list(subjects.keys())

    for a in slots:
        for b in slots:
            for c in slots:
                plan = {s[0]: a, s[1]: b, s[2]: c}
                plans.append(plan)
    return plans


# UNIT-II: Heuristic Function (Informed Search)
def heuristic(plan):
    score = 0
    for subject, slot in plan.items():
        if slot == "Morning":
            score += subjects[subject]
    return score


# UNIT-III: Constraint Satisfaction
def valid_plan(plan):
    return len(set(plan.values())) == len(plan.values())


# Main Execution
plans = generate_plans(subjects, time_slots)
valid_plans = [p for p in plans if valid_plan(p)]
best_plan = max(valid_plans, key=heuristic)

print("Best Study Plan Suggested by AI:\n")
for subject, slot in best_plan.items():
    print(subject, "->", slot)
```
